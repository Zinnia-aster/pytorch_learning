###Matrix multiplication two main rules which will prevent shape errors.

1. The inner dimensions must match:

(3, 2) @ (3, 2) won’t work

(2, 3) @ (3, 2) will work

(3, 2) @ (2, 3) will work

2. The resulting matrix has the shape of the outer dimensions:

(2, 3) @ (3, 2) → (2, 2)

(3, 2) @ (2, 3) → (3, 3)


```python
torch.rand(2,5)
```

```python
tensor = torch.rand(2,5)@torch.rand(5,2)
tensor
```

```python
tensor.shape
```

When we have two tensors in the same shape but we need to do matrix multiplication means it does not follow the rules to do matrix multiplication. In that case we use transpose. We transpose one of the tensor and then do matrix multiplication then it will work.

```python
tensor_a = torch.tensor([[2,4],
                        [4,7],
                        [5,3]])
tensor_b = torch.tensor([[4,6],
                        [3,7],
                        [2,4]])


ttensor = tensor_a@tensor_b.T
ttensor,ttensor.shape
```

###Reshaping, stacking, squeezing, and unsqueezing tensors.


*   Reshaping- Reshapes an input tensor to a defined shape,
*  view- to return a view of an input tensor of certain shape but keeps the same memory as the original tensor.
* Stacking- Combines multiple tensors on top of each other(vstack) or side-by-side (hstack).
* Squeeze- Removes all one dimensions from a tensor.
* Unsqueeze- Add a one dimension to a target tensor.
*  Permute- Return a view of the input with dimensions permuted, swapped in a certain way.




```python
x = torch.arange(1.,10.)
x,x.shape
```

the reshaping value must be compatible, like when the shape of the tensor is like 9, it has to be multiplied, the reshaping value has to be multiplied like 3 into 3,1 into 9 or 9 into 1. If we are not giving a compatible values, then it won't work.

```python
x_reshaped = x.reshape(1,9)
x_reshaped
```

```python
x_reshaped = x.reshape(3,3)
x_reshaped
```

```python
# changing the view
z = x.view(1,9)
z,z.shape
```

```python
# changing z changes x(because a view of a tensor shares a same memory as the original tensor)
z[:,3] = 8
x,z
```

##stacking
stacking can be done in three diff ways

```python
#1
x_stacked = torch.vstack((x,x,x,x))    #vertical stack
x_stacked
```

OR

```python
x_stacked = torch.stack((x,x,x,x),dim=0)
x_stacked
```

```python
#2
x_stacked = torch.stack((x,x,x,x),dim=1)
x_stacked
```

```python
#3
x_stacked = torch.hstack((x,x,x,x))        #horizontal stack
x_stacked
```

###squeezing and unsqueezing

```python
#squeezing
print(f"previous tensor:{x_reshaped}")
print(f"previous shape:{x_reshaped.shape}")

#remove 1 dimension tensor
x_squeezed = x.squeeze()

print(f"\nnew tensor:{x_squeezed}")
print(f"new shape:{x_squeezed.shape}")
```

```python
#unsqueezing
print(f"previous tensor:{x_squeezed}")
print(f"previous shape:{x_squeezed.shape}")

#add an extra dimension
x_unsqueezed = x_squeezed.unsqueeze(dim=0)

print(f"\nnew tensor:{x_unsqueezed}")
print(f"new shape:{x_unsqueezed.shape}")
```

##reproduciblity

 when we use a random values we use a random seed to flavor the randomness
 which gives reproduciblity (able to use again and again and again)

```python
tensor_a = torch.rand(3,4)
tensor_b = torch.rand(3,4)

print(tensor_a)
print(tensor_b)
print(tensor_a == tensor_b)
```

```python
#now using random seed method
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
tensor_c = torch.rand(3,4)
torch.manual_seed(RANDOM_SEED)
tensor_d = torch.rand(3,4)

print(tensor_c)
print(tensor_d)
print(tensor_c == tensor_d)
```
